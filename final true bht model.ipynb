```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv('/Halliburton AI - ML Challenge/Codes/Datasets/final dummy data.csv')
```

```python
df
```

```python
X = df.drop(columns=['true bht'])
```

```python
y = df['true bht']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.ensemble import BaggingRegressor
from sklearn.metrics import r2_score,accuracy_score,confusion_matrix,precision_score

reg = BaggingRegressor(n_estimators=100)
```

```python
reg.fit(X_train,y_train)
y_pred = reg.predict(X_test)
accuracy = r2_score(y_test,y_pred)

print("For ",reg)
print("Accuracy - ",accuracy)
```

```python
import pickle
pickle.dump(reg, open('final-true-bht.pkl', 'wb'))
```
